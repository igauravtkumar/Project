```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['Total Sales'], color='skyblue')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Total Sales'], color='blue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_method = df['Payment Method'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(payment_method, labels=payment_method.index, autopct='%1.1f%%', startangle=140, colors=['Red', 'Violet', 'Pink','Skyblue','Blue'])
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.countplot(data=df, x='Status', order=df['Status'].value_counts().index, palette='pastel')
plt.title("Order Status Count")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()
```

```python
top_locations = df.groupby('Customer Location')['Total Sales'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(data=top_locations, x='Customer Location', y='Total Sales', palette='Blues_d')
plt.title("Top 10 Customer Locations by Sales")
plt.xlabel("Customer Location")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
